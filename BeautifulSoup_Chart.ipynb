```python
URL = "https://www.statista.com/chart/20903/countries-emitting-most-co2-per-capita/"
r = requests.get(url=URL, headers=headers)
  
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
```

```python
soup.find_all('img')
```

```python
mydivs = soup.find_all("div", {"class": "article__graphic"})
mydivs
```

```python
class_=str(mydivs[0])
```

```python
url = class_[class_.find("data-src")+10:class_.find('height="1200"')-2]
```

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get(url)
img = Image.open(BytesIO(response.content))
img
```
